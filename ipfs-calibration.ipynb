# Test: Size of window
Compare transition flows across two periods to see which is more stable
## add parameters (7 day window, 14 day window)

```python
params1 = {
    'start_date': '2021-01-18',
    'end_date': '2021-02-20',
    'threshold': 5,
    'window': 7,
    'dependency_threshold': 1
}
```

```python
params2 = {
    'start_date': '2021-01-18',
    'end_date': '2021-02-20',
    'threshold': 10,
    'window': 14,
    'dependency_threshold': 1
}
```

```python
ipfs_window_data1 = json.loads(requests.get('https://ipfs.ecosystem-dashboard.com/repositories/states.json', params=params1).content)
filecoin_window_data1 = json.loads(requests.get('https://filecoin.ecosystem-dashboard.com/repositories/states.json', params=params1).content)
ipfs_window_data2 = json.loads(requests.get('https://ipfs.ecosystem-dashboard.com/repositories/states.json', params=params2).content)
filecoin_window_data2 = json.loads(requests.get('https://filecoin.ecosystem-dashboard.com/repositories/states.json', params=params2).content)
```

```python
earliest_date1 = min([d['date'] for d in ipfs_window_data1])
print(earliest_date1)
earliest_date2 = min([d['date'] for d in ipfs_window_data2])
print(earliest_date2)
```

```python
# get list of users by state for the latest date
ipfs_base1 = ipfs_window_data1[[x1['date'] for x1 in ipfs_window_data1].index(earliest_date1)]['states']
filecoin_base1 = filecoin_window_data1[[x1['date'] for x1 in filecoin_window_data1].index(earliest_date1)]['states']
# create combined high value list for tracking
base1 = list(dict.fromkeys([r['repo_name'] for r in ipfs_base1['high']] + \
                               [r['repo_name'] for r in filecoin_base1['high']]))

# get list of users by state for the latest date
ipfs_base2 = ipfs_window_data2[[x1['date'] for x1 in ipfs_window_data2].index(earliest_date2)]['states']
filecoin_base2 = filecoin_window_data2[[x1['date'] for x1 in filecoin_window_data2].index(earliest_date2)]['states']
# create combined high value list for tracking
base2 = list(dict.fromkeys([r['repo_name'] for r in ipfs_base2['high']] + \
                               [r['repo_name'] for r in filecoin_base2['high']]))
```

```python
# who appears in the first set of params that doesn't in the second?
set(base1) - set(base2)
```

```python
# who appears in the second set of params that doesn't in the first?
set(base2) - set(base1)
```

```python
# first get for IPFS
transitions1 = {
    u: {
        data['date']:state 
            for data in ipfs_window_data1 
            for state, user_data in data['states'].items()
            for entry in user_data
            if u == entry['repo_name']
    }
for u in base1}
# then for filecoin
second_results = {
    u: {
        data['date']:state 
            for data in filecoin_window_data1 
            for state, user_data in data['states'].items()
            for entry in user_data
            if u == entry['repo_name']
    }
for u in base1}
# then combine
for u in base1:
    if transitions1[u] == {}:
        transitions1[u] = second_results[u]

# first get for IPFS
transitions2 = {
    u: {
        data['date']:state 
            for data in ipfs_window_data2
            for state, user_data in data['states'].items()
            for entry in user_data
            if u == entry['repo_name']
    }
for u in base2}
# then for filecoin
second_results = {
    u: {
        data['date']:state 
            for data in filecoin_window_data2
            for state, user_data in data['states'].items()
            for entry in user_data
            if u == entry['repo_name']
    }
for u in base2}
# then combine
for u in base2:
    if transitions2[u] == {}:
        transitions2[u] = second_results[u]
```

```python
def calc_changes(transitions):
    first = True
    for repo, value in transitions.items():
        if first:
            num_dates = len(value)
            tracker = [0] * (num_dates + 1)
            first = False
        dum = 0
        for date, state in value.items():
            if state == 'high':
                dum += 1
        tracker[dum] += 1
    return tracker
```

```python
change1 = calc_changes(transitions1)
perc_by_num_transitions_high1 = [c / sum(change1) for c in change1]
print(change1)
print(perc_by_num_transitions_high1)
print('Share of repos that were high value in at least 3 of 4 weeks: ', perc_by_num_transitions_high1[3] + perc_by_num_transitions_high1[4])
```

```python
change2 = calc_changes(transitions2)
perc_by_num_transitions_high2 = [c / sum(change2) for c in change2]
print(change2)
print(perc_by_num_transitions_high2)
print('Share of repos that were high value in both two week periods: ', perc_by_num_transitions_high2[2])
```

So a 14 day window seems right
