# Test: Size of window
Compare transition flows across two periods to see which is more stable
## add parameters (7 day window, 14 day window)

```python
params1 = {
    'start_date': '2021-02-07',
    'end_date': '2021-02-20',
    'threshold': 5,
    'window': 7,
    'dependency_threshold': 1
}
```

```python
params2 = {
    'start_date': '2021-02-07',
    'end_date': '2021-02-20',
    'threshold': 10,
    'window': 14,
    'dependency_threshold': 1
}
```

```python
data1 = json.loads(requests.get('https://ipfs.ecosystem-dashboard.com/repositories/transitions.json', params=params1).content)
data2 = json.loads(requests.get('https://ipfs.ecosystem-dashboard.com/repositories/transitions.json', params=params2).content)
```

```python
data1[0]['transitions'].keys()
```

## transitions in/out of high value list and differences

```python
transitioned_in1 = [r['repo_name'] for r in data1[0]['transitions']['Low to High-Value']] + \
    [r['repo_name'] for r in data1[0]['transitions']['Reactivated High-Value']]
for u in sorted(transitioned_in1):
    print(u)
```

```python
transitioned_out1 = [r['repo_name'] for r in data1[0]['transitions']['High to Low-Value']] + \
    [r['repo_name'] for r in data1[0]['transitions']['Lapsed High-Value']]
for u in sorted(transitioned_out1):
    print(u)
```

```python
# do any users just cycle in / out?
list(set(transitioned_in1) & set(transitioned_out1))
```

```python
transitioned_in2 = [r['repo_name'] for r in data2[0]['transitions']['Low to High-Value']] + \
    [r['repo_name'] for r in data2[0]['transitions']['Reactivated High-Value']]
for u in sorted(transitioned_in1):
    print(u)
```

```python
transitioned_out2 = [r['repo_name'] for r in data2[0]['transitions']['High to Low-Value']] + \
    [r['repo_name'] for r in data2[0]['transitions']['Lapsed High-Value']]
for u in sorted(transitioned_out2):
    print(u)
```

```python
# do any users just cycle in / out?
list(set(transitioned_in2) & set(transitioned_out2))
```

```python
# who does list 1 have transitioning in that's not in list 2 that we like / don't like?
list(set(transitioned_in1) - set(transitioned_in2))
```

```python
# who does list 2 have transitioning in that's not in list 1 that we like / don't like?
list(set(transitioned_in2) - set(transitioned_in1))
```

```python
# who does list 1 have transitioning out that's not in list 2 that we like / don't like?
list(set(transitioned_out1) - set(transitioned_out2))
```

```python
# who does list 2 have transitioning out that's not in list 1 that we like / don't like?
list(set(transitioned_out2) - set(transitioned_out1))
```
